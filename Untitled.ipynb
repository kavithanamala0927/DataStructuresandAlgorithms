```python
from collections import Counter
def check(s):
    o = Counter(s)
    r = Counter(o.values()).most_common()[0][0]
    count = 0  
    d = {}
    flag = 0
    for k,v in o.items():
        if o[k] != r:
            d[k] = v
            flag = 1
    if flag == 1:
        c = sum(d.values())-1
        if c == 0 or c==r:
            return ("YES")
        elif c != r:
            return ("NO")
    else:
        return ("YES")
s = "aabbcde"
check(s)
```

```python
from itertools import combinations
def substrCount(n, s):
    l = (s[i:j] for i,j in combinations(range(n+1), 2))
    count = 0
    for i in l:
        if i == i[::-1]:
            count += 1
    return count
string = """"""
#string = "aaaa"
n = 727310
#n = 4
#1272919 
substrCount(n, string)
```

```python
def substrCount(n, s):
    count = n
    for x in range(n): 
        y = x
        while y < n - 1:
            y += 1
            if s[x] == s[y]:
                count += 1
            else:
                if s[x:y] == s[y+1:y+2]:
                    count += 1
                break

    return count
```

```python
import numpy as np
import math
def calculate_efficiency_from_current(measured_current, cosphi):
        load_factor = calculate_load_current_ratio(measured_current)
        return calculate_efficiency(load_factor, cosphi)

def calculate_load_current_ratio(measured_current):
        nominal_current = 145
        load_current_ratio = float(measured_current) / float(nominal_current)
        return load_current_ratio
def calculate_efficiency(loading_factor, cosphi):
        """
        This method calculates efficiency of the transformer based on the
        inputs load_factor & cosphi.
        This calculation includes calculation of voltage drop as a first step
        :param loading_factor:
        :param cosphi: load angle of the transformer
        :return: returns efficiency of the transformer rounded to 2 digits
        """
        p0_w = 4900
        pk_120w = 37250
        sn_kva = 2.5*1000
        voltage_drop = calculate_voltage_drop(loading_factor, cosphi)
        numerator = (p0_w / 1000) + (math.pow(loading_factor, 2) *
                                                      (pk_120w / 1000))
        denominator = loading_factor * sn_kva * cosphi * (1 - voltage_drop / 100)
        denominator += numerator
        return np.round(100 - (numerator * 100 / denominator), 2)
    
def calculate_voltage_drop(loading_factor, cosphi):
        """
        This method calculates voltage_drop of the transformer based on the
        inputs load_factor & cosphi
        :param loading_factor:
        :param cosphi: load angle of the transformer
        :return: returns voltage_drop of the transformer rounded to 2 digits
        """
        p0_w = 4900
        pk_120w = 37250
        sn_kva = 2.5*1000
        uz = 9.8
        ur = np.round((pk_120w * 0.1) / sn_kva, 4)
        ux = np.round(math.sqrt(math.pow(uz, 2) - math.pow(ur, 2)), 2)
        result1 = loading_factor * (ur * cosphi +
                                    ux * math.sqrt(1 - math.pow(cosphi, 2)))
        result2 = loading_factor * (ur * math.sqrt(1 - math.pow(cosphi, 2)) -
                                    ux * cosphi)
        return np.round(result1 + (math.pow(result2, 2) / 200), 2)
def calculate_max_optimum_current_efficiency(cosphi):
    efficiencies = []
    loading_factors = np.arange(0.05,1.00,0.05)
    for loading_factor in loading_factors:
        efficiency = calculate_efficiency(loading_factor, cosphi)
        efficiencies.append(efficiency)
    max_optimum_current_efficiency = 1.0005 * max(efficiencies)
    return np.round(max_optimum_current_efficiency, 2)
    
actual_efficiency = calculate_efficiency_from_current(145, 0.5)
print (actual_efficiency)
target_efficiency = calculate_max_optimum_current_efficiency(0.5)
print (target_efficiency)
```
